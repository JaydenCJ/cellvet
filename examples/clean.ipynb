# Quarterly revenue analysis (reproducible)

Same analysis, written to survive *Restart & Run All*.

```python
import statistics
```

```python
revenue = [1200, 1350, 980, 1420]
tax_rate = 0.25
```

```python
mean_revenue = statistics.mean(revenue)
mean_revenue
```

```python
total = mean_revenue * (1 + tax_rate)
round(total, 2)
```
